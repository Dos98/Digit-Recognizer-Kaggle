```python
import numpy
import matplotlib.pyplot
%matplotlib inline
```

```python
# open the CSV file and read its contents into a list
data_file = open("/home/ayush/.kaggle/competitions/digit-recognizer/train.csv", 'r')
data_list = data_file.readlines()
data_file.close()
```

```python
# check the number of data records (examples)
len(data_list)
```

```python
# show a dataset record
# the first number is the label, the rest are pixel colour values (greyscale 0-255)
data_list[1]
```

```python
# take the data from a record, rearrange it into a 28*28 array and plot it as an image
#load test data
test_data_file = open("/home/ayush/.kaggle/competitions/digit-recognizer/test.csv", 'r')
test_data_list = test_data_file.readlines()
test_data_file.close()
for record in test_data_list[1:20]:
    all_values = record.split(',')
    image_array = numpy.asfarray(all_values[:]).reshape((28,28))
    matplotlib.pyplot.figure()
    matplotlib.pyplot.imshow(image_array, cmap='Greys', interpolation='None')
```

```python
# scale input to range 0.01 to 1.00
scaled_input = (numpy.asfarray(all_values[:]) / 255.0 * 0.99) + 0.01
print(scaled_input)
```

```python
img = numpy.asfarray(scaled_input[:]).reshape((28,28))
matplotlib.pyplot.imshow(img, cmap='Greys', interpolation='None')
```
